# 2.Check Stationary State with Dickey-Fuller Augmented Test
Stationary state shows data points does not move over time. If Data is non-stationary, better to detrend and/or drift data set to stationary state.
Results show that all data is stationary. Confident level is 0.95. p-values all less than 0.05

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(stockname):
    timeseries=volf[stockname].dropna()
    print(stockname)
    print ('Results of Dickey-Fuller Test: ')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
       dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
    print(" ")
for i in volf:
    adf_test(i)
```

Rationale for VAR selection


The information criteria for optimal lag length  is contingent on the number of observations. While the AIC and Final Prediction Error (FPE) are more appropriate when observations are less than 60, the Hannan-Quin is more efficient when observations are above 120. Moreover, it remains the discretion of the researcher to select the maximum lags  which the adopted criterion for choosing optimal lags will use.
Reference
Liew, V, K-S., (2004). "What lag selection criteria should we employ?", Economics Bulletin, 33(3), pp. 1-9.

## a.Static Full Sample Period From 20.07.2007 to 20.04.2020
The fevd the variance decompostion can be easily transferred to the generalized fevd by multiplying sigma u and premultiplying the fevd with standard deviation of u.
VAR(3). And forcast horizon of H= 10

### Table Matrix

```python
sigma_u = py.asarray(results.sigma_u)
sd_u = py.sqrt(py.diag(sigma_u))
fevd=results.fevd(10, sigma_u / sd_u)
#fevd.summary()
fe = fevd.decomp[:,-1,:]
gfevd=fe/fe.sum(1)[:,None]*100

#Net Pairwise Connectedness: DONE
transpose=gfevd.T
diff=(transpose-gfevd)
net_connect = py.tril(diff) + py.tril(diff, -1).T
mat=pd.DataFrame(data=net_connect,index=names, columns=names)
mat=mat.replace(0,1)

#Create table OutPUT:DONE
con_to = gfevd.sum(0) - py.diag(gfevd)
con_from = gfevd.sum(1) - py.diag(gfevd)
con_to=py.append(con_to,(gfevd.sum()-py.trace(gfevd))/10)
exp=pd.DataFrame(data=py.round(gfevd,3), index=names, columns=names)
exp['FROM']=py.round(con_from,3)
exp.loc['TO']=py.round(con_to,3)
exp
```
